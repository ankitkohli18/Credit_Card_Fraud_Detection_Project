```python
#importing required libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
#reading the dataset
cc = pd.read_csv('/content/creditcard.csv', error_bad_lines=False)
```

```python
cc.head()
```

Since, credit card details are vey sensitive and they cannot be exposed so, principal component analysis is performed on those features

```python
cc.tail()
```

```python
cc.info()
```

```python
cc.isnull().sum()
```

```python
cc['Class'].value_counts()
```

This dataset is highly unbalanced

0 --> legit transaction

1 --> Fraudulent transaction

```python
legit = cc[cc.Class==0]
fraudulent = cc[cc.Class==1]
```

```python
print(legit.shape)
print(fraudulent.shape)
```

```python
legit.Amount.describe()
```

```python
fraudulent.Amount.describe()
```

```python
cc.groupby('Class').mean()
```

This dataset is highly unbalanced so, we need to perform undersampling inorder to balance the dataset

```python
legit_sample = legit.sample(n = 492)
```


Merging legit_sample and fraudulent transactions dataframe

```python
new_dataset = pd.concat([legit_sample, fraudulent], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Splitting dataset into features(X) and targets(Y)
X = new_dataset.drop(columns='Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training Model using

Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model evalutation

Accuracy Score

```python
X_train_prediction = model.predict(X_train)
training_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on training data: ', training_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on testing data: ', testing_accuracy)
```
